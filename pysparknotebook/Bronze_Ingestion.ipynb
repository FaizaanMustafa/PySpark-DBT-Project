```python
entities = ['customers','drivers','locations','payments','trips','vehicles']
```

```python
for entity in entities:

    df_entity  = spark.read.format('csv').option('header','true').option('inferSchema','true').option('path',f'/Volumes/pysparkdbt/source/source_data/{entity}/').load()

    schema_entity = df_entity.schema

    df = spark.readStream.format('csv').option('header','true').schema(schema_entity).load(f'/Volumes/pysparkdbt/source/source_data/{entity}')
    
    df.writeStream.format('delta').outputMode('append').option('checkpointLocation',f'/Volumes/pysparkdbt/bronze/checkpoint/{entity}').trigger(once=True).table(f'pysparkdbt.bronze.{entity}')
```
